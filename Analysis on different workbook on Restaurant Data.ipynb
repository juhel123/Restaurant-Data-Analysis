#### **WHAT ARE THE TOP 5 FOOD ITEMS THAT WERE SOLD**

```python
import pandas as pd
import matplotlib.pyplot as plt

foodsdf= pd.read_csv("./Restaurant Datas/Restaurant - Foods.csv")
week1sales= pd.read_csv("./Restaurant Datas/Restaurant - Week 1 Sales.csv")
week2sales= pd.read_csv("./Restaurant Datas/Restaurant - Week 2 Sales.csv")
customersdf= pd.read_csv("./Restaurant Datas/Restaurant - Customers.csv")
foodsdf
week2sales
customersdf
week1sales
```

```python
week1_satisfactiondf= pd.read_csv("./Restaurant Datas/Restaurant - Week 1 Satisfaction.csv")

week1_satisfactiondf
```

```python
foodsales= foodsdf.groupby("Food Item")["Price"].sum().reset_index(name="Prices of Food Items")

foodsales
```

```python
top5sales= foodsales.sort_values(by="Prices of Food Items", ascending= False).head(5)
top5sales
```

#### **WHO IS THE CUSTOMER THAT ORDERED THE MOST IN THE WEEK2**

```python
# note that in order to find the customer that ordered the most, we will find this customer by name and food id bcos food id will give us the number of times each customer placed an order, so in order to do this we will first convert the ID in consumers workbook to customers ID in week2 sales then merge the result to week2 workbooks
customersdf.rename(columns={"ID" : "Customer ID"}, inplace=True)
merged_week2customer= pd.merge(week2sales,customersdf, on="Customer ID", how="inner")

merged_week2customer
```

```python
Customers_order= merged_week2customer.groupby("First Name")["Food ID"].sum().reset_index(name="Customers order")

Customers_order
```

```python
most_ordered_customer= Customers_order.sort_values(by="Customers order", ascending=False).head(1)

most_ordered_customer
```

```python
most_ordered_customer = Customers_order.nlargest(1, 'Customers order')

most_ordered_customer
```

#### **WHAT IS THE OCCUPATION OF THE CUSTOMER WHO ORDERED THE LEAST AMOUNT OF FOOD IN THE WEEK1 AND WEEK2**

#### **WEEK 1 CUSTOMER ORDER BY NAME AND OCCUPATION**

```python
customersdf.rename(columns={"ID" : "Customer ID"}, inplace=True)
merged_week1customer= pd.merge(week1sales,customersdf, on="Customer ID", how="inner")

merged_week1customer
```

```python
Customers_orders= merged_week1customer.groupby("First Name")["Food ID"].sum().reset_index(name="No of times customer placed an order in week 1")

Customers_orders
```

```python
Customers_orders= merged_week1customer.groupby("Occupation")["Food ID"].sum().reset_index(name="No of times customer placed an order in week 1")

Customers_orders
```

```python
week1_least_amount_of_order =  Customers_orders.loc[Customers_orders['No of times customer placed an order in week 1'].idxmin()]

print(f"The occupation of the customer that ordered the least amount of food is {week1_least_amount_of_order["Occupation"]} and he/she placed {week1_least_amount_of_order["No of times customer placed an order in week 1"]} order")
```

#### **WEEK 2 CUSTOMER ORDER BY NAME AND OCCUPATION**

### **BASED ON PRICE, LIST THE TOP SEVEN MOST EXPENSIVE FOODS**

### **WHAT ARE THE NUMBER OF ORDERS PLACED FROM EACH WEEK IE WEEK1 AND WEEK2 OR HOW MANY ORDERS WERE PLACED FROM EACH WEEK**

```python
week1_order= week1sales.groupby("Customer ID")["Food ID"].sum().reset_index()

week1_order
```

```python
#Since each order may have multiple items (and thus multiple rows with the same food_id), you need to count the unique food_id values to determine the total number of orders:
Total_week1_order = week1_order['Food ID'].nunique()

print(f"The total number of order placed in week 1 is {Total_week1_order}")
```
